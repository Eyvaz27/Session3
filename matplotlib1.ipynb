### Works related to matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sns.set()
# %matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
function1 = np.sin(x)
function2 = np.cos(x)
plt.plot(x, function1, '-m')    # matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, **kwargs)
plt.plot(x, function2, ':r')
plt.legend(['sin', 'cos'])
plt.xlabel('argument')
plt.ylabel('sin')
plt.title('sin function')
```

```python
random = (np.random.standard_normal(100))
l = []
for i in random:
    l.append(np.int(i))
print(l)
x = range(0, len(l))
print(random.std())
print(random.mean())
plt.hist(random, bins = 50)
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.random.randint(0, 30, 100)
plt.scatter(x, y, label = 'scatter', marker = '^', color = 'm', linewidths = 1)
```

```python
days = [1, 2, 3, 4, 5, 6,  7]
sleeping = [7, 8, 6, 5, 6, 8, 10]
eating = [1, 2, 1.3, 1.5, 1, 2, 4]
playing = [1, 1.6, 1.3, 1.7, 1.5, 2, 7]
working = [15, 12.4, 15.4, 15.8, 15.5, 12, 3]
cols = ['m', 'b', 'c', 'r']

plt.plot([], [], color = 'm', label = 'sleeping', linewidth = 3)
plt.plot([], [], color = 'b', label = 'eating', linewidth = 3)
plt.plot([], [], color = 'c', label = 'playing', linewidth = 3)
plt.plot([], [], color = 'r', label = 'working', linewidth = 3)
plt.stackplot(days, sleeping, eating, playing, working, colors = cols)
plt.legend()
plt.xlabel('days')
plt.ylabel('activities')
plt.title('Stack Plot')
plt.show()
```

```python
slices = [7, 2, 2, 13]
activities = ['sleeping', 'eating', 'working', 'playing']
cols = ['m', 'b', 'c', 'r']
plt.pie(slices, labels = activities, 
        colors = cols, shadow = True, radius = 1, startangle = 90, 
        counterclock = False, explode = (0.1, 0.1, 0.1, 0.1), autopct = '%.2f%%')
```
