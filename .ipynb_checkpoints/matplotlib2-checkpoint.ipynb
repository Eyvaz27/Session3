### Works related to Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sns.set()
```

```python
x, y, z = np.loadtxt('example.txt', dtype = np.int, delimiter = ',', unpack = True)
print(x, y, z)
y = np.sin(y)
z = np.cos(z)
plt.plot(x, y, label = 'x-sin(y)', linewidth = 2, color = 'm', marker = 's', linestyle = '-')
plt.plot(x, z, label = 'x-cos(z)', linewidth = 2, color = 'r', marker = '^', linestyle = ':')
plt.legend()
plt.show()
```

#### Importing Data from Internet

```python
import urllib
import matplotlib.dates as mdates
```

```python
def bytesdate2num(fmt, encoding = 'utf-8'):
    function = mdates.strpdate2num(fmt)
    def bytes_convert(b):
        s = b.decode(encoding)
        return function(s)
    return bytes_convert

def graph_data(stock):
    
    stock_price_url = 'https://pythonprogramming.net/yahoo_finance_replacement'
    stock_code = urllib.request.urlopen(stock_price_url).read().decode()
    stock_data = []
    stock_source = stock_code.split('\n')
    
    for line in stock_source[1:]:
        stock_data.append(line)
    
    (date, closep, highp, lowp, openp, adj_closep, volume) = np.loadtxt(stock_data, delimiter = ',', unpack = True,
                                                            converters={0: bytesdate2num('%Y-%m-%d')})
    
    plt.figure(num = 1, figsize=(15, 4))
    plt.subplot(131)
    plt.plot_date(date, openp, label = 'open rate', linewidth = 2, linestyle = '-', markersize = 0.5, color = 'm')
    plt.plot_date(date, closep, label = 'close rate', linewidth = 2, linestyle = ':', markersize = 0.1, color = 'c')
    plt.legend()
    plt.subplot(132)
    plt.plot_date(date, highp, label = 'high price', linewidth = 2, linestyle = '-', markersize = 0.5, color = 'g')
    plt.plot_date(date, lowp, label = 'low price', linewidth = 2, linestyle = ':', markersize = 0.1, color = 'k')
    plt.legend()
    plt.subplot(133)
    plt.title('Total Volume')
    plt.plot_date(date, volume, label = 'volume', linewidth = 2, linestyle = '-.', markersize = 1, color = 'r')
    plt.legend()
    plt.show()
graph_data('Tesla')
```
